```python
#import libraries

import pandas as pd
import seaborn as sns
```

```python
#import data 
ultra = pd.read_csv('./data_practice/TWO_CENTURIES_OF_UM_RACES.csv')

#open it 
ultra
```

```python
ultra.dtypes
```

```python
#show if df has 50 miles or 50 kilometers listed under event distance/length column
ultra[ultra['Event distance/length'] == '50km']
```

```python
ultra[ultra['Event distance/length'] == '50m']
```

```python
ultra[ultra['Event distance/length'] == '50mi']
```

```python
#combine 50k/50mi with isin
ultra[ultra['Event distance/length'].isin(['50km','50mi'])]
```

```python
#add & to filter for 2018 year only 
ultra[(ultra['Event distance/length'].isin(['50km', '50mi'])) & (ultra['Year of event'] == 2018)]
```

```python
#filter even further with CHILE events only and create a new df called chile
chile_ultras = ultra[(ultra['Event distance/length'].isin(['50km', '50mi'])) & 
                  (ultra['Year of event'] == 2018) & 
                  (ultra['Event name'].str.contains('(CHI)'))]

chile_ultras
```

```python
#remove CHI from event name 
chile_ultras['Event name'] = chile_ultras['Event name'].str.split('(').str.get(0)
```

```python
chile_ultras.head()
```

```python
#clean up athlete age by crating a new column and name it Athelet Age
#will get age by subtracting year of event (2018) by athlete year of birth

chile_ultras['Athlete Age'] = 2018 - chile_ultras['Athlete year of birth']

chile_ultras
```

```python
#remove h from athlete performance since it's redundant 
chile_ultras['Athlete performance'] = chile_ultras['Athlete performance'].str.split(' ').str.get(0)

chile_ultras
```

```python
#drop columns: Athlete Club, Athlete Country, Athlete year of birth, Athlete Age Category

chile_ultras = chile_ultras.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'], axis=1)

chile_ultras
```

```python
#cleanup null values 

#find total nan values for each column
chile_ultras.isna().sum()
```

```python
#find rows
chile_ultras[chile_ultras['Athlete Age'].isna()==1]
```

```python
#drop rows
chile_ultras = chile_ultras.dropna()
```

```python
chile_ultras.shape
```

```python
#check for dupes
```

```python
chile_ultras[chile_ultras.duplicated() == True]
```

```python
#reset index

chile_ultras.reset_index(drop = True)
```

```python
#fix athlete age from float to int & athlete avg speed from obj to float
```

```python
chile_ultras.dtypes
```

```python
chile_ultras['Athlete Age'] = chile_ultras['Athlete Age'].astype(int)
```

```python
chile_ultras.dtypes
```

```python
chile_ultras['Athlete average speed'] = chile_ultras['Athlete average speed'].astype(float)
```

```python
chile_ultras.dtypes
```

```python
chile_ultras.head()
```

```python
#rename columns to remove spaces and make everything lowercase
chile_ultras = chile_ultras.rename(columns = {'Year of event':'year',
                                              'Event dates': 'race_day',
                                              'Event distance/length': 'race_lenght',
                                              'Event name': 'race_name',
                                              'Event number of finishers': 'race_number_of_finishers',
                                              'Athlete performance': 'athlete_performance',
                                              'Athlete gender': 'athlete_gender',
                                              'Athlete average speed': 'athlete_average_speed',
                                              'Athlete ID': 'athlete_id',
                                              'Athlete Age': 'athlete_age'
                                             })

chile_ultras.head()
```

```python
#reorder columns and not include year column since it is redundant

new_chile_ultras = chile_ultras[['race_day','race_name','race_lenght','race_number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performance', 'athlete_average_speed']]

new_chile_ultras
```

```python
#create a graph that shows how many ultra races are listed in km vs mi
sns.histplot(new_chile_ultras['race_lenght'])
```

```python
#create another race that breaks it down by male and female gender
sns.histplot(new_chile_ultras, x = 'race_lenght', hue = 'athlete_gender')
```

```python
#create a plot chart that shows the distribution count of athlete avg speed 
sns.displot(new_chile_ultras[new_chile_ultras['race_lenght'] == '50km']['athlete_average_speed'])
```

Looks like the average are between 6km and 7km

```python
#graph to observe the average speed for male and female athletes
sns.violinplot(data=new_chile_ultras, x='race_lenght', y='athlete_average_speed', hue='athlete_gender', split=True, inner='quartiles')
```

```python
#scatter plot to compare it with age instead and look for outliers 
sns.lmplot(data=new_chile_ultras, x = 'athlete_age', y='athlete_average_speed')
```

```python
#difference in speed for the 50km male to female 

new_chile_ultras.groupby(['race_lenght','athlete_gender'])['athlete_average_speed'].mean()
```

```python
#what age groups are the best in the 50km race (20 + race minimum)

new_chile_ultras[new_chile_ultras['race_lenght'] == '50km'].groupby('athlete_age').agg(
    mean_speed=('athlete_average_speed', 'mean'),
    race_count=('athlete_id', 'count')
).reset_index()
```
